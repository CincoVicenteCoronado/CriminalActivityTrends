```python
import pandas as pd
import gzip
```

## Exploración inicial de los datos

```python
# Lee el archivo CSV
df = pd.read_csv('Chicago_Crimes_2008_to_2011.csv')
```

```python
df.head()
```

```python
## Le ponemos un nuevo nombre a la columna nueva
df = df.rename(columns={'Unnamed: 0': 'indice'})
```

```python
# verificación de nulos
nulos_por_columna = df.isna().sum()
print(nulos_por_columna)
```

Encontramos que la columna Case Number tienen pocos nulos por lo que borraremos las filas con Case Number nulo de df

```python
df[df['Case Number'].isna()==True]
```

```python
ids = [629323, 652024, 1013929, 1776819, 1799520, 2161425]
df = df.drop(ids)
```

```python

## borraremos las columnas con nulos para limpiar el dataset
df = df.dropna(axis=1)
```

```python
nulos_por_columna = df.isna().sum()
print(nulos_por_columna)
```

## Ordenaremos por la fecha

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')

# Ordenar el DataFrame por la columna 'Date'
df = df.sort_values(by='Date')

# Mostrar el DataFrame ordenado
df.head()
```

```python
# Guarda el DataFrame como un archivo TSV sin índice
df.to_csv('archivo.tsv', sep='\t', index=False)

# Comprime el archivo TSV a .gz
with open('archivo.tsv', 'rb') as f_in:
    with gzip.open('archivo.tsv.gz', 'wb') as f_out:
        f_out.writelines(f_in)
```
